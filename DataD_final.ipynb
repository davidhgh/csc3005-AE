 __Remove outliers__

```python
import numpy as np
from scipy import stats
dataD[(np.abs(stats.zscore(dataD)) < 3).all(axis=1)]
```

```python
dataD_noclass = dataD.drop('HealthState', axis=1)
X=dataD_noclass.to_numpy()
Y=dataD[['HealthState']].to_numpy()
Y=Y.reshape(-1)

X
```

__PCA EigenDecomposition__

```python
from sklearn.decomposition import PCA
pca_all = PCA()
pca_all.fit(X)
```

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
eigen_energy=pca_all.explained_variance_ratio_
print(eigen_energy)
plt.bar(range(1,44), eigen_energy, alpha=0.5, align='center', label='individual explained variance')
plt.xlabel('PC')
plt.ylabel('eigen_variance_energy_ratio')
plt.title("Plot of eigen variance energy captured by Principle component")
```

__Based on the plot, we should use at least 2 PCs as they covered at least 90% energy__
