 __Remove outliers__

```python
import numpy as np
from scipy import stats
pd.set_option('display.max_rows', None)
dataD = dataD[(np.abs(stats.zscore(dataD)) < 3).all(axis=1)]
classGrp = dataD.groupby('HealthState')
classGrp.count()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

dataD_noclass = dataD.drop('HealthState', axis=1)
dataD_noclass_scaled = scaler.fit_transform(dataD_noclass)
X= dataD_noclass_scaled
Y=dataD[['HealthState']].to_numpy()
Y=Y.reshape(-1)

X
```

__PCA EigenDecomposition__

```python
from sklearn.decomposition import PCA
pca_all = PCA()
pca_all.fit(X)
```

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
eigen_energy=pca_all.explained_variance_ratio_
print(eigen_energy)
plt.bar(range(1,44), eigen_energy, alpha=0.5, align='center', label='individual explained variance')
plt.xlabel('PC')
plt.ylabel('eigen_variance_energy_ratio')
plt.title("Plot of eigen variance energy captured by Principle component (Data D)")
```

__Based on the plot, we should use at least 2 PCs as they covered at least 90% energy__

__Very hard to decide which 2 features to use, if all feature to be used, use 3 PC instead?__

```python
pca_2 = PCA(n_components=2)
pca_2.fit(X)
X_pca_2=pca_2.transform(X)
print(X_pca_2)
print(X_pca_2.shape) #only 2 PC used
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X_pca_2, Y, test_size=0.3, random_state=12)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
%matplotlib inline

numNeighbors = [1, 5, 6,7,8, 10, 15, 20, 25, 30]
trainAcc = []
testAcc = []

for k in numNeighbors:
    clf1 = KNeighborsClassifier(n_neighbors=k, metric='minkowski', p=2)
    clf1.fit(X_train, Y_train)
    Y_predTrain = clf1.predict(X_train)
    Y_predTest = clf1.predict(X_test)
    trainAcc.append(accuracy_score(Y_train, Y_predTrain))
    testAcc.append(accuracy_score(Y_test, Y_predTest))

plt.plot(numNeighbors, trainAcc, 'ro-', numNeighbors, testAcc,'bv--')
plt.legend(['Training Accuracy','Test Accuracy'])
plt.xlabel('Number of neighbors')
plt.ylabel('Accuracy')
```
