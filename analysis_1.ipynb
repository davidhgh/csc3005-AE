```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score

# KNN
from sklearn.neighbors import KNeighborsClassifier

# Naive Bayes
from sklearn.naive_bayes import GaussianNB

# SVC
from sklearn.svm import SVC


dataB2_df = dataB_df
dataB2_df['HealthState'] = dataB['HealthState']
dataB2_df
```

```python
def usingSVC(df, colSize):
    # Split data
    X = df.iloc[:,0:colSize].values
    y = df.iloc[:,colSize].values
    # split dataset 80 20
    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, test_size=0.2)
    # split dataset 70 30
    X_train_2, X_test_2, y_train_2, y_test_2 = train_test_split(X, y, random_state=0, test_size=0.3)

    print("Split dataset 80:20")
    print(f"No. of training dataset (X): {len(X_train)}, No. of testing dataset: {len(X_test)}")
    print(f"No. of training dataset (y): {len(y_train)}, No. of testing dataset: {len(y_test)}")
    print()
    print("Split dataset 70:30")
    print(f"No. of training dataset (X): {len(X_train_2)}, No. of testing dataset: {len(X_test_2)}")
    print(f"No. of training dataset (y): {len(y_train_2)}, No. of testing dataset: {len(y_test_2)}")

    ## SVC 
    #Feature scaling
    sc_X = StandardScaler()

    # Split dataset 80:20
    X_train_scale = sc_X.fit_transform(X_train)
    X_test_scale = sc_X.transform(X_test)
    # Split dataset 70 30
    X_train_2_scale = sc_X.fit_transform(X_train_2)
    X_test_2_scale = sc_X.transform(X_test_2)


    ## SVC 
    svc1 = SVC() 

    # Without Feature scaling
    y_pred = svc1.fit(X_train,y_train).predict(X_test)
    y_pred_2 = svc1.fit(X_train_2,y_train_2).predict(X_test_2)

    # With Feature scaling
    y_pred_scale = svc1.fit(X_train_scale,y_train).predict(X_test_scale)
    y_pred_2_scale = svc1.fit(X_train_2_scale,y_train_2).predict(X_test_2_scale)
    print()
    print("===============================================================")
    print("|                             SVC                             |")
    print("===============================================================")

    print("=====================Without Feature scaling===================")
    print("Split dataset 80:20")
    print("Number of mislabeled points out of a total %d points : %d"% (X_test.shape[0], (y_test != y_pred).sum()))
    print("Accuracy score: " + str(accuracy_score(y_test, y_pred)))
    print()
    print("Split dataset 70:30")
    print("Number of mislabeled points out of a total %d points : %d"% (X_test_2.shape[0], (y_test_2 != y_pred_2).sum()))
    print("Accuracy score: " + str(accuracy_score(y_test_2, y_pred_2)))

    print("======================With Feature scaling=====================")
    # With Feature scaling
    print("Split dataset 80:20")
    print("Number of mislabeled points out of a total %d points : %d"% (X_test_scale.shape[0], (y_test != y_pred_scale).sum()))
    print("Accuracy score: " + str(accuracy_score(y_test, y_pred_scale)))
    print()
    print("Split dataset 70:30")
    print("Number of mislabeled points out of a total %d points : %d"% (X_test_2_scale.shape[0], (y_test_2 != y_pred_2_scale).sum()))
    print("Accuracy score: " + str(accuracy_score(y_test_2, y_pred_2_scale)))

usingSVC(dataB2_df, 12)
```

```python
def useKNN(df, colSize):
    # Split data
    X = df.iloc[:,0:colSize].values
    y = df.iloc[:,colSize].values
    # split dataset 80 20
    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, test_size=0.2)
    # split dataset 70 30
    X_train_2, X_test_2, y_train_2, y_test_2 = train_test_split(X, y, random_state=0, test_size=0.3)

    print("Split dataset 80:20")
    print(f"No. of training dataset (X): {len(X_train)}, No. of testing dataset: {len(X_test)}")
    print(f"No. of training dataset (y): {len(y_train)}, No. of testing dataset: {len(y_test)}")
    print()
    print("Split dataset 70:30")
    print(f"No. of training dataset (X): {len(X_train_2)}, No. of testing dataset: {len(X_test_2)}")
    print(f"No. of training dataset (y): {len(y_train_2)}, No. of testing dataset: {len(y_test_2)}")

    ## KNN 
    #Feature scaling
    sc_X = StandardScaler()

    # Split dataset 80:20
    X_train_scale = sc_X.fit_transform(X_train)
    X_test_scale = sc_X.transform(X_test)
    # Split dataset 70 30
    X_train_2_scale = sc_X.fit_transform(X_train_2)
    X_test_2_scale = sc_X.transform(X_test_2)


    # Define the model: Init K-NN
    classifier = KNeighborsClassifier(n_neighbors=11, p=2,metric='euclidean')

    # Without Feature scaling
    y_pred = classifier.fit(X_train,y_train).predict(X_test)
    y_pred_2 = classifier.fit(X_train_2,y_train_2).predict(X_test_2)

    # With Feature scaling
    y_pred_scale = classifier.fit(X_train_scale,y_train).predict(X_test_scale)
    y_pred_2_scale = classifier.fit(X_train_2_scale,y_train_2).predict(X_test_2_scale)

    print("===============================================================")
    print("|                             KNN                             |")
    print("===============================================================")

    print("=====================Without Feature scaling===================")
    print("Split dataset 80:20")
    print("Number of mislabeled points out of a total %d points : %d"% (X_test.shape[0], (y_test != y_pred).sum()))
    print("Accuracy score: " + str(accuracy_score(y_test, y_pred)))
    print()
    print("Split dataset 70:30")
    print("Number of mislabeled points out of a total %d points : %d"% (X_test_2.shape[0], (y_test_2 != y_pred_2).sum()))
    print("Accuracy score: " + str(accuracy_score(y_test_2, y_pred_2)))

    print("======================With Feature scaling=====================")
    # With Feature scaling
    print("Split dataset 80:20")
    print("Number of mislabeled points out of a total %d points : %d"% (X_test_scale.shape[0], (y_test != y_pred_scale).sum()))
    print("Accuracy score: " + str(accuracy_score(y_test, y_pred_scale)))
    print()
    print("Split dataset 70:30")
    print("Number of mislabeled points out of a total %d points : %d"% (X_test_2_scale.shape[0], (y_test_2 != y_pred_2_scale).sum()))
    print("Accuracy score: " + str(accuracy_score(y_test_2, y_pred_2_scale)))

useKNN(dataB2_df, 12)
```
