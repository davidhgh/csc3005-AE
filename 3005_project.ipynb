```python
dataC = dataC.drop(['FlowVelocityP1', 'FlowVelocityP2','FlowVelocityP3','FlowVelocityP4', 'SoundSpeedP1', 'SoundSpeedP2','SoundSpeedP3','SoundSpeedP4'], axis=1)
dataC
```

```python
# Remove any duplicated data, if any
dataD.drop_duplicates()
```

```python
# Drop all Flow velocity as there is already an average velocityFlow of 4 path.
# Drop all sound_of_speed as there is already an average sound_of_speed of 4 path.
# Drop all signalStrength as there is already signalQuality of throughput.
# Drop all transit time as the speed of sound is derived from: 
#       sos = (sum of transitTime/ product) * 1/2 of path length
#       futhermore, our group is already using the avg sos.

dataD_aggregated = dataD.drop(['FlowVelocityP1', 'FlowVelocityP2', 'FlowVelocityP3', 'FlowVelocityP4',
                    'SoundSpeedP1', 'SoundSpeedP2', 'SoundSpeedP3', 'SoundSpeedP4',
                    'SignalStrengthP1_1', 'SignalStrengthP1_2', 'SignalStrengthP2_1', 'SignalStrengthP2_2', 'SignalStrengthP3_1', 'SignalStrengthP3_2', 'SignalStrengthP4_1', 'SignalStrengthP4_2',
                    'TransmitTimeP1_1', 'TransmitTimeP1_2', 'TransmitTimeP2_1', 'TransmitTimeP2_2', 'TransmitTimeP3_1', 'TransmitTimeP3_2', 'TransmitTimeP4_1', 'TransmitTimeP4_2'], axis=1)
dataD_aggregated
```

```python
# Remove all dataset that have signal quality = 0

# REASON:
# USM electronic reject unstable signal quality, making data 0
# (reference: https://asgmt.com/wp-content/uploads/pdf-docs/2011/1/M15.pdf, page 2, section "Signal Quality - Transducer Performance")

# Since signal quality is inaccurate, data will not be useful in this case.
dataD_aggregated.drop(dataD_aggregated[dataD_aggregated['SignalQualityP1_1'] == 0].index, inplace=True)
dataD_aggregated.drop(dataD_aggregated[dataD_aggregated['SignalQualityP1_2'] == 0].index, inplace=True)
dataD_aggregated.drop(dataD_aggregated[dataD_aggregated['SignalQualityP2_1'] == 0].index, inplace=True)
dataD_aggregated.drop(dataD_aggregated[dataD_aggregated['SignalQualityP2_2'] == 0].index, inplace=True)
dataD_aggregated.drop(dataD_aggregated[dataD_aggregated['SignalQualityP3_1'] == 0].index, inplace=True)
dataD_aggregated.drop(dataD_aggregated[dataD_aggregated['SignalQualityP3_2'] == 0].index, inplace=True)
dataD_aggregated.drop(dataD_aggregated[dataD_aggregated['SignalQualityP4_1'] == 0].index, inplace=True)
dataD_aggregated.drop(dataD_aggregated[dataD_aggregated['SignalQualityP4_2'] == 0].index, inplace=True)
 
dataD_aggregated
```

```python
from sklearn.decomposition import PCA

dataD_aggregated_no_class = dataD_aggregated.drop(["HealthState"], axis=1)

numComponents_D = 15
pca_d = PCA(n_components=numComponents_D)
pca_d.fit(dataD_aggregated_no_class)
projected_d = pca_d.transform(dataD_aggregated_no_class)
dataD_aggregated_pca = pd.DataFrame(projected_d)
dataD_aggregated_pca["HealthState"] = dataD_aggregated["HealthState"]

dataD_aggregated_pca
```

```python
import matplotlib.pyplot as plt
import numpy as np

# calculate cumulative sum of explained variances
tot = sum(pca_d.explained_variance_)
print(pca_d.explained_variance_)
print(tot)
var_exp = [(i / tot) for i in sorted(pca_d.explained_variance_, reverse=True)]
cum_var_exp = np.cumsum(var_exp)
print(len(var_exp))
# plot explained variances
plt.bar(range(1,numComponents_D + 1), var_exp, alpha=0.5,
        align='center', label='individual explained variance')
plt.step(range(1,numComponents_D + 1), cum_var_exp, where='mid',
         label='cumulative explained variance')
plt.ylabel('Explained variance ratio')
plt.xlabel('Principal component index')
plt.legend(loc='best')
plt.show()

```

```python
plt.plot(np.cumsum(pca_d.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance');
```

```python
# Class '1' - Healthy
# Class '2' - Gas injection
# Class '3' - Waxing

colors = {1: 'g', 2: 'b', 3: 'r'}
markerTypes = {1: '+', 2: 'x', 3: 'o'}

for healthStateType in markerTypes:
    # print(healthStateType)
    d = dataD_aggregated_pca[dataD_aggregated_pca['HealthState'] == healthStateType]
    plt.scatter(d[0], d[1], c=colors[healthStateType], s=200, marker=markerTypes[healthStateType])
```
