```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

numComponents = 2
pca = PCA(n_components=numComponents)
pca.fit(dataB)
projected = pca.transform(dataB)
projected_1 = pd.DataFrame(projected)

pca.fit(dataC)
projected = pca.transform(dataC)
projected_2 = pd.DataFrame(projected)

pca.fit(dataD)
projected = pca.transform(dataD)
projected_3 = pd.DataFrame(projected)

projected_1.style
projected_2.style
projected_3.style
```

```python
plt.figure(0)
plt.title("Set B (PCA 2 Components)")
plt.scatter(projected_1[0],projected_1[1])

plt.figure(1)
plt.title("Set C (PCA 2 Components)")
plt.scatter(projected_2[0],projected_2[1])

plt.figure(2)
plt.title("Set D (PCA 2 Components)")
plt.scatter(projected_3[0],projected_3[1])
```
